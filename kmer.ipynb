```python
from collections import Counter
from Bio import SeqIO
from Bio.SeqUtils import lcc
```

```python
records = SeqIO.parse('/media/GenomicResearch/Issue/20211026_沙鹿童綜合Sal/Complete_genome/5A-sal-05_1.fa', 'fasta')
record = next(records)
```

```python
seq = record.seq
```

```python
k = 31
```

```python
stop = len(seq) - k + 1
```

```python
c = Counter(seq[i:i+k] for i in range(stop))
```

```python
c.most_common(1)
```

```python
kmer_complexity = [(kmer, lcc.lcc_simp(kmer)) for kmer in c.keys()]
```

```python
sorted(kmer_complexity, key=lambda x: x[1], reverse=True)
```
