```python
import os
import subprocess
from pathlib import Path
from concurrent.futures import ProcessPoolExecutor
```

```python
def run_command(cmd):
    subprocess.run(cmd, shell=True, check=True)
```

```python
dirpath = Path('/media/NGS/SRA_1/NCBI_Listeria_monocytogenes_SRA/Contigs')
outpath = Path('/media/NGS/SRA_1/NCBI_Listeria_monocytogenes_SRA/KMA/results')
database = "/media/NGS/SRA_1/NCBI_Listeria_monocytogenes_SRA/KMA/kma_db"
```

```python
with ProcessPoolExecutor(64) as executor:
    for i in dirpath.iterdir():
        outfile =outpath/i.stem
        cmd = f"kma -i {i} -o {outfile} -t_db {database} -Sparse"
        executor.submit(run_command, cmd)
```

```python
import pandas as pd
```

```python
dirpath = Path('/media/NGS/SRA_1/NCBI_Listeria_monocytogenes_SRA/KMA/results')
```

```python
profiles = []
for i in dirpath.iterdir():
    profile = pd.read_csv(i, sep='\t')
    try:
        profile.index = [i.stem]
        profiles.append(profile)
    except ValueError:
        pass
```

```python
df = pd.concat(profiles)
```

```python
df.to_csv('/media/NGS/SRA_1/NCBI_Listeria_monocytogenes_SRA/KMA.tsv', sep='\t')
```
