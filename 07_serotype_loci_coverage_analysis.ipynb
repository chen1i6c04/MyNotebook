```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

```python
coverage = pd.read_csv("../results/serotype_loci_coverage.tsv", sep="\t", index_col=0).T
coverage
```

```python
num_strain = 7449
coverage = coverage.apply(lambda x: x * 100 / num_strain)
coverage["level"] = list(map(lambda x: np.floor(x/2)*2, coverage.sum(1)))
```

```python
def hist(df):
    count = df.shape[0]
    return count * df.sum() / df.sum().sum()
```

```python
perc = coverage.groupby("level").agg(hist)
perc
```

```python
perc["Total"] = perc.sum(1)
perc
```

```python
fig = plt.figure(figsize=(16, 12))
plt.plot(perc.index[2:], perc["Total"].iloc[2:], drawstyle="steps")
plt.xlabel("Percentage of genomes covered by loci (%)", fontsize=18)
plt.ylabel("Number of locus", fontsize=18)
# plt.show()
fig.savefig("../pics/genome_loci_coverage.png")
```

```python
pair = ("Typhi", "Total")

fig = plt.figure(figsize=(16, 12))
plt.plot(perc.index[2:], perc[pair[0]].iloc[2:], drawstyle="steps", label=pair[0])
plt.plot(perc.index[2:], perc[pair[1]].iloc[2:], drawstyle="steps", label=pair[1])
plt.title("Serotype loci coverage (5% <= x < 100%): {}".format(pair[0]), fontsize=22)
plt.legend(prop={'size': 16})
plt.xlabel("Percentage of genomes covered by loci (%)", fontsize=18)
plt.ylabel("Number of locus", fontsize=18)
# plt.show()
fig.savefig("../pics/serotype_loci_coverage_{}.png".format(pair[0]))
```
