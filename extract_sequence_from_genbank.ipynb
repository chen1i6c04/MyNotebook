```python
import pandas as pd
from Bio import SeqIO
```

```python
filepath = "/home/chen1i6c04/下載/sequence.gb"
```

```python
records = SeqIO.parse(filepath, 'genbank')
```

```python
record = next(records)
```

```python
chrom = 'HF937208.1'
positions = []
for feature in record.features:
    convertor = {1: '+', -1: '-'}
    if feature.type == "rRNA":
        locus_tag = dict(feature.qualifiers)['locus_tag'][0]
        strand = feature.strand
        start, end = feature.location.start.position, feature.location.end.position
        positions.append((chrom, start, end, locus_tag, convertor[strand]))
```

```python
df = pd.DataFrame(positions)
```

```python
df.to_csv('sequences.bed', sep='\t', index=False, header=False)
```
