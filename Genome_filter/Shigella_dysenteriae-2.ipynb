```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats.mstats import gmean
```

```python
dirpath = '/media/NGS/Sequence/Shigella_dysenteriae/quast'
```

```python
reports = []
for dirname in os.listdir(dirpath):
    report = pd.read_csv(os.path.join(dirpath, dirname, 'report.tsv'), sep='\t', index_col=0)
    reports.append(report)
```

```python
table = pd.concat(reports, axis=1).T
```

```python
identify = pd.read_csv('/media/NGS/Sequence/Shigella_dysenteriae/Identify.csv', index_col=0)
identify = identify[(identify['Query_Coverage']>=50)&(identify['Template_Coverage']>=50)&(identify['Species']=='Shigella dysenteriae')]
```

```python
table = table.filter(identify.index, axis=0)
```

```python
assembly_info = pd.read_csv('/media/NGS/Sequence/Shigella_dysenteriae/prokaryotes.csv')

complete = assembly_info[(assembly_info['Level'] == 'Complete')|(assembly_info['Level'] == 'Chromosome')]['Assembly'].to_list()
```

```python
u_table = table[~table.index.isin(complete)]
c_table = table[table.index.isin(complete)]
```

```python
u_table = u_table[u_table['N50']>=gmean(u_table['N50'])]
```

```python
u_table.sort_values('N50', ascending=False)
```

```python
table = pd.concat([c_table, u_table], axis=0)
```

```python
def fn(x):
    mean = np.mean(x)
    std = np.std(x)
    return (x < mean + (2 * std))&((mean - 2*std) < x)
```

```python
table[fn(table['Total length (>= 0 bp)'])&fn(table['GC (%)'])]
```
