```python
import sys
from pathlib import Path
import pandas as pd

sys.path.append('../src')
from tree import Dendrogram, PairwiseDistanceMatrix
```

```python
dirpath_1 = Path('/media/GenomicResearch/Issue/20210803_genomic_distance_with_nanopore/Contigs')
dirpath_2 = Path('/media/GenomicResearch/MiSeq/Salmonella_enterica/Profile')
```

```python
accs = set(map(lambda x: x.stem, dirpath_1.iterdir()))
```

```python
profile = pd.concat([pd.read_csv(dirpath_2/(acc + '.tsv'), sep='\t', index_col=0, header=0, names=[acc]) for acc in accs], axis=1)
```

```python
distmatrix = PairwiseDistanceMatrix(profile)()
```

```python
dendrogram = Dendrogram(distmatrix)
dendrogram()
```

```python
dendrogram.show_figure()
```

```python
dendrogram.to_newick('/media/GenomicResearch/Issue/20210803_genomic_distance_with_nanopore/cgmlst.newick')
```
