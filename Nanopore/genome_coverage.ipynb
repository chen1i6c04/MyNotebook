```python
from collections import defaultdict
import pandas as pd
from Bio import SeqIO
```

```python
template_file = '/media/NGS/Nanopore_1/mNGS/20210121_spike/untrim/unicycler/assembly.fasta'
```

```python
template_size = sum(len(record.seq) for record in SeqIO.parse(template_file, 'fasta'))
```

```python
names = ["Query","Q_length","Q_start","Q_end","Strand","Template","T_length","T_start",
         "T_end","N_res_matches","Align_block",]
```

```python
df = pd.read_csv('/media/NGS/Nanopore_1/mNGS/20210121_spike/untrim/overlap/barcode12.paf', sep='\t', names=names, usecols=range(0, 11))
```

```python
cover_range = defaultdict(set)
```

```python
for template, start, end in zip(df['Template'], df['T_start'], df['T_end']):
    cover_range[template].update(list(range(start, end+1)))
```

```python
cover_size = sum(len(value) for value in cover_range.values())
```

```python
round(cover_size/template_size*100, 2)
```
