```python
import subprocess
from pathlib import Path

import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from Bio import SeqIO
```

```python
fig = plt.figure(dpi=150)

fastq_file = '/media/NGS/Nanopore_1/mNGS/20210121_spike/untrim/Rawreads/barcode09.fastq'
reads_len = [len(record.seq) for record in SeqIO.parse(fastq_file, 'fastq')]
maxvalx = max(reads_len)

hist_kws = {'linewidth': 1, 'histtype': 'step', 'color': 'b', 'alpha': 1}
sns.distplot(a=np.log10(reads_len), kde=False, hist=True, bins=max(round(int(maxvalx) / 1000), 10),
             hist_kws=hist_kws, label='barcode09')

fastq_file = '/media/NGS/Nanopore_1/mNGS/20210121_spike/untrim/Rawreads/barcode10.fastq'
reads_len = [len(record.seq) for record in SeqIO.parse(fastq_file, 'fastq')]
maxvalx = max(reads_len)

hist_kws = {'linewidth': 1, 'histtype': 'step', 'color': 'r', 'alpha': 1}
sns.distplot(a=np.log10(reads_len), kde=False, hist=True, bins=max(round(int(maxvalx) / 1000), 10),
             hist_kws=hist_kws, label='barcode10')

fastq_file = '/media/NGS/Nanopore_1/mNGS/20210121_spike/untrim/Rawreads/barcode11.fastq'
reads_len = [len(record.seq) for record in SeqIO.parse(fastq_file, 'fastq')]
maxvalx = max(reads_len)

hist_kws = {'linewidth': 1, 'histtype': 'step', 'color': 'g', 'alpha': 1}
sns.distplot(a=np.log10(reads_len), kde=False, hist=True, bins=max(round(int(maxvalx) / 1000), 10),
             hist_kws=hist_kws, label='barcode11')

fastq_file = '/media/NGS/Nanopore_1/mNGS/20210121_spike/untrim/Rawreads/barcode12.fastq'
reads_len = [len(record.seq) for record in SeqIO.parse(fastq_file, 'fastq')]
maxvalx = max(reads_len)

hist_kws = {'linewidth': 1, 'histtype': 'step', 'color': 'black', 'alpha': 1}
sns.distplot(a=np.log10(reads_len), kde=False, hist=True, bins=max(round(int(maxvalx) / 1000), 10),
             hist_kws=hist_kws, label='barcode12')

plt.legend(loc=2)

ticks = [10**i for i in range(10) if not 10**i > 10 * maxvalx]
fig.axes[0].set(
    xticks=np.log10(ticks),
    xticklabels=ticks,
    xlabel='Read length',
);
```
