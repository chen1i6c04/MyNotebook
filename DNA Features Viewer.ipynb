```python
from dna_features_viewer import GraphicFeature, GraphicRecord, BiopythonTranslator
from Bio import SeqIO
import matplotlib.pyplot as plt
```

```python
color_map = {
    "rep_origin": "yellow",
    "CDS": "orange",
    "regulatory": "red",
    "misc_recomb": "darkblue",
    "misc_feature": "lightblue",
}

translator = BiopythonTranslator(
    features_filters=(lambda f: f.type not in ["gene", "source"] and 'product' in f.qualifiers.keys(),),
    features_properties=lambda f: {"color": color_map.get(f.type, "white")},
)
```

```python
def fn(gbk_file, translator, window, plot_sequence=False):
    graphic_record = translator.translate_record(gbk_file)
    cropped_record = graphic_record.crop(window)
    ax, _ = cropped_record.plot(figure_width=20, strand_in_label_threshold=10)
    if plot_sequence:
        cropped_record.plot_sequence(ax=ax)
```

```python
fn("/media/Access/3_個人儲存區/陳柏翰/TVGH/TVGH20001-0318.gbk", translator, (1835155, 1835185), True)
```
