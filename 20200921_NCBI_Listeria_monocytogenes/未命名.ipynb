```python
import pandas as pd
```

```python
run_info = pd.read_csv('/media/NGS/Data_Analysis/NCBI/Run_Info/CDC_Listeria_monocytogenesSraRunInfo.csv', usecols=['Run', 'BioSample'])
sample_attribute = pd.read_csv('/media/NGS/Data_Analysis/NCBI/Sample_Info/CDC_Listeria_monocytogenes_SampleInfo.csv', dtype=object)
cgtype = pd.read_csv('/media/Synology_222/3_個人儲存區/陳柏翰/CDC_lmonocytogenes_ctype.csv', dtype=object)
amr = pd.read_csv('/media/NGS/Data_Analysis/20191128_CDC_MiSeq/Listeria_monocytogenes/AMR.csv')
```

```python
sample_attribute = sample_attribute.merge(run_info).drop(['host age', 'host sex', 'isolate', 'collected by'], axis=1)
```

```python
sample_attribute = sample_attribute.merge(cgtype, left_on='strain', right_on='Key').rename(columns={'CT': 'cgMLST'}).drop('Key', axis=1)
```

```python
sample_attribute = sample_attribute.merge(amr, left_on='strain', right_on='Key').rename(columns={'AMR genes': 'ResistGene'}).drop('Key', axis=1)
```

```python
rename_map = {'strain': 'Identifer',
              'collection date': 'Year',
              'geographic location': 'Country',
              'host': 'Host',
              'isolation source': 'SampleSource',
              'subtype': 'ST',
              'serotype': 'Serovar',
              'Run': 'DNASource'}
```

```python
sample_attribute = sample_attribute.rename(rename_map, axis=1)
```
