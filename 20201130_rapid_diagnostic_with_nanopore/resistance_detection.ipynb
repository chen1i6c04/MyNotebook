```python
import pandas as pd
```

```python
filepath = "/media/NGS/Nanopore_1/mNGS/20210209_clBC/resfinder.tsv"
correct_resistance_genes = pd.read_csv(filepath, sep='\t', index_col=0)
correct_resistance_genes = correct_resistance_genes.apply(lambda x: x.dropna().str.cat(sep=', ').split(', '), axis=1).to_dict()
```

```python
filepath = "/media/NGS/Nanopore_1/mNGS/20210209_clBC/time_filt/resfinder.tsv"
time_filter_resistance_genes = pd.read_csv(filepath, sep='\t', index_col=0)
time_filter_resistance_genes = time_filter_resistance_genes.apply(lambda x: x.dropna().str.cat(sep=', ').split(', '), axis=1).to_dict()
```

```python
data = []
for sample, genes in time_filter_resistance_genes.items():
    sample_name, time = sample.split('_')
    overlap = set(correct_resistance_genes[sample_name]) & set(genes)
    data.append((sample_name, len(overlap), len(set(correct_resistance_genes[sample_name])), time.replace('h', '')))
```

```python
df = pd.DataFrame(data, columns=['sample', 'match', 'total', 'hour'])
df
```

```python
df.to_csv('/media/NGS/Nanopore_1/mNGS/20210209_clBC/time_filt.tsv', sep='\t', index=False)
```
